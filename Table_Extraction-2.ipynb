```python
import tabula

def extract_table_titles(pdf_path):
    # Read the PDF file and extract tables
    tables = tabula.read_pdf(pdf_path, pages='all', multiple_tables=True)

    # Iterate through tables and extract titles
    for i, table in enumerate(tables):
        # Extract the first line of the table as the title
        title = table.iloc[0, 0] if not table.empty else None

        # Print or use the title
        if title:
            print(f"Table {i + 1} Title: {title}")
        else:
            print(f"No title found for Table {i + 1}.")

# Example usage
pdf_path = "/content/drive/MyDrive/Audited_Financial_Statements/Documents_1/2023_Jan.pdf"
extract_table_titles(pdf_path)
```

```python
def remove_extra_spaces(text):
    # Replace multiple spaces with a single space
    cleaned_text = re.sub(r'\s+', ' ', text)
    # Trim leading and trailing spaces
    cleaned_text = cleaned_text.strip()
    return cleaned_text
```
