```python
import openai
import fitz  # PyMuPDF
import re


# Specify pages to read
pages_to_read = [7, 8, 9, 10]

# Read text from specific pages of the PDF
text_from_pages = []
with fitz.open(pdf_file_path) as pdf_document:
    for page_number in pages_to_read:
        page = pdf_document.load_page(page_number - 1)
        text = page.get_text("text")
        cleaned_text = clean_text(text)
        cleaned_text = remove_extra_spaces(cleaned_text)
        text_from_pages.append(cleaned_text)

# Concatenate the text into a single string
combined_text = "\n".join(text_from_pages)

# Instruction for summarization
instruction = "Please provide a summary of the Income Statement, Balance Sheet, and Cash Flow Statement from the document. Additionally, calculate key financial ratios including Earnings per Share (EPS), Price-to-Earnings (P/E) Ratio, and Return on Equity (ROE). Identify and highlight any significant changes in important financial metrics that may influence stock price changes in the future."

# Concatenate the text and instruction
full_prompt = combined_text + "\n" + instruction

# Make the API request
response = openai.Completion.create(
    engine="gpt-3.5-turbo-instruct",
    prompt=full_prompt,
    max_tokens=2000  # Adjust based on the desired length of the summary
)

# Extract and print the summary from the response
summary = response['choices'][0]['text']

# Print the summary without the instruction
print("Summary:", re.sub(re.escape(instruction), '', summary).strip())
```
